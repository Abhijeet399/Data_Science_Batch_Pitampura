```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
```

```python
print(themes)
plt.style.use('bmh')
```

```python
x = np.arange(10)
```

```python
x
```

```python
y1 = x**2
y2 = 2*x +12
```

```python
plt.plot(x, y1, color = 'red', label='Oil', linestyle='--')
plt.plot(x, y2, c = 'green', label = 'Gold')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Time vs Price graph')
plt.legend(title='Prices:')
plt.show()
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize = (10,10))
plt.scatter(x, y1, c = 'blue', label='scatter_plot', marker='.')
plt.scatter(x, y2, c = 'green', label='scatter_plot', marker='+')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Time vs Price graph')
plt.legend(title='Prices:')
plt.show()
```

```python
plt.bar(['zero', 'one', 'two'], [10, 20, 15])
plt.show()
```

```python
x_cor = np.array([0,1,2])
plt.bar(x_cor, [10, 20, 15], width = 0.4, color = 'green', label='sales', tick_label=["a", "b", "c"])
plt.bar(x_cor + 0.5, [50, 10, 35], width = 0.4)
plt.ylim(0, 60)
plt.xlim(-0.5, 3)
plt.show()
```
